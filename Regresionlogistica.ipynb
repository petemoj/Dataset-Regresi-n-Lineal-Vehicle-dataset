```python
#Revisión de datos faltantes

print("Cantidad de Registros con cero en cp: ",Datos['cp'].isin([0]).sum())
print("Cantidad de Registros con cero en trestbps: ",Datos['trestbps'].isin([0]).sum())
print("Cantidad de Registros con cero en chol: ",Datos['chol'].isin([0]).sum())
print("Cantidad de Registros con cero en fbs: ",Datos['fbs'].isin([0]).sum())
print("Cantidad de Registros con cero en restecg: ",Datos['restecg'].isin([0]).sum())
```

```python
# Reemplazar todos los ceros con el valor medio (promedio) de la columna.

Datos['cp']=Datos['cp'].replace(0,Datos['cp'].mean())
Datos['trestbps']=Datos['trestbps'].replace(0,Datos['trestbps'].mean())
Datos['chol']=Datos['chol'].replace(0,Datos['chol'].mean())
Datos['fbs']=Datos['fbs'].replace(0,Datos['fbs'].mean())
Datos['restecg']=Datos['restecg'].replace(0,Datos['restecg'].mean())
```

```python
#Revisión de datos faltantes

print("Cantidad de Registros con cero en cp: ",Datos['cp'].isin([0]).sum())
print("Cantidad de Registros con cero en trestbps: ",Datos['trestbps'].isin([0]).sum())
print("Cantidad de Registros con cero en chol: ",Datos['chol'].isin([0]).sum())
print("Cantidad de Registros con cero en fbs: ",Datos['fbs'].isin([0]).sum())
print("Cantidad de Registros con cero en restecg: ",Datos['restecg'].isin([0]).sum())
```

```python
#Conteo de Datos Atípicos

nivel_minimo=0
nivel_maximo=3
Total_Atipicos=((Datos['cp']<nivel_minimo)|(Datos['cp']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))

nivel_minimo=94
nivel_maximo=200
Total_Atipicos=((Datos['trestbps']<nivel_minimo)|(Datos['trestbps']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))

nivel_minimo=126
nivel_maximo=564
Total_Atipicos=((Datos['chol']<nivel_minimo)|(Datos['chol']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))

nivel_minimo=0
nivel_maximo=1
Total_Atipicos=((Datos['fbs']<nivel_minimo)|(Datos['fbs']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))

nivel_minimo=0
nivel_maximo=2
Total_Atipicos=((Datos['fbs']<nivel_minimo)|(Datos['fbs']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
#Característica de interes:
#Condition: Si la persona tiene enfermedad cardiaca o no (0 = no tiene enfermedad cardiaca - 1 = si tiene enfermedad cardiaca
Datos['condition'].value_counts()
```

```python
Datos[['condition']].head(40)
```

```python
X = Datos.drop('condition',axis=1) # Se cargan los datos de train sin los datos de la columna condition
Y = Datos['condition'] # Se cargan los datos de train con los datos de la la columna condition

#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state= 0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

```python
# Construir el modelo de regresión logística 

modelo_regresion = LogisticRegression(solver = "liblinear") # liblinear es una libreria para resolver problemas lineales de clasificación y regresión
modelo_regresion.fit (X_train , Y_train)   

# Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test) # la función predict predice los valores a partir del conjunto de datos de validación
```

```python
print (Y_test)
```

```python
print (Prediccion)
```

```python
# Desempeño del modelo de regresión

print(classification_report(Y_test,Prediccion))
```

```python
#Matrix de Confusión del modelo

cm=confusion_matrix(Y_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisón del modelo

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, Prediccion)
plt.show()
```
